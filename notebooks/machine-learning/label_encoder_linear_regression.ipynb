```python
#Step 1: import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
import sys
import os

#Step 2: set data path for local/Colab
if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive', force_remount=True)
    data_path = '/content/drive/My Drive/Projects/GenerativeAI/data/data.csv'
else:
    data_path = os.path.abspath(os.path.join('..','..','data','data.csv'))

#Step 3: read data
df = pd.read_csv(data_path)

#Step 4: apply label encoding
label_encoder = LabelEncoder()
encoded_job = label_encoder.fit_transform(df['Job'])

#Step 5: Prepare the target
X = encoded_job.reshape(-1, 1)
y = df['Age']

#Step 6: Create a model
model = LinearRegression()
model.fit(X, y)
```
