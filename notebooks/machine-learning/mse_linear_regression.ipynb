```python
#Step 1: import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import sys

# Set data path based on environment
if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive')
    data_path = '/content/drive/My Drive/Projects/GenerativeAI/data/'
else:
    data_path = '../../data/'

#Step 2: read the data
df = pd.read_csv(data_path + 'Sales.csv')

#Step 3: prepare the features
X = df[['Advertising Expenditure']]
y = df['Sales']

#Step 4: model
model = LinearRegression()
model.fit(X,y)

#Step 5: Predictions
predictions = model.predict(X)

#Step 6: Calculate MSE
mse = mean_squared_error(y, predictions)
```
