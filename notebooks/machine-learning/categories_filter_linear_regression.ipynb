```python
#Step 1: import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import sys
import os

#Step 2: set data path for local/Colab
if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive', force_remount=True)
    data_path = '/content/drive/My Drive/Projects/GenerativeAI/data/data.csv'
else:
    data_path = os.path.abspath(os.path.join('..','..','data','data.csv'))

#Step 3: read data
df = pd.read_csv(data_path)

#Step 4: filtered data
filtered_df = df[(df['Sex'] == 'male') & (df['Job'] == 'skilled')]

#Step 5: Prepare data
X = filtered_df[['Sex', 'Job']]
y = filtered_df['Age']

#Step 6: convert categorical features
X_encoded = pd.get_dummies(X)

#Step 7: linear regression
model = LinearRegression()
model.fit(X_encoded, y)
```
