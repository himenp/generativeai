```python
# === BOOTSTRAP: Find project root & setup sys.path ===
import sys
from pathlib import Path

print("=" * 60)
print("BOOTSTRAP")
print("=" * 60)

def find_project_root(markers=("requirements.txt", "config.yaml", ".git")):
    cwd = Path.cwd().resolve()
    for path in [cwd] + list(cwd.parents):
        if any((path / marker).exists() for marker in markers):
            return path
    raise FileNotFoundError("Project root not found")

PROJECT_ROOT = find_project_root()
print(f"Project root: {PROJECT_ROOT}")

if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
    print(f"✓ Added to sys.path")

print("=" * 60)
```

```python
# === LOAD CONFIGURATION ===
import yaml

print("=" * 60)
print("LOAD CONFIG")
print("=" * 60)

config_path = PROJECT_ROOT / "config" / "config.yaml"
if not config_path.exists():
    raise FileNotFoundError(f"Config not found at {config_path}")

with open(config_path, "r", encoding="utf-8") as f:
    config = yaml.safe_load(f) or {}

project_name = config.get('project_name', 'GenerativeAI')
print(f"Project name: {project_name}")
print("✓ Config loaded")
print("=" * 60)
```

```python
# === IMPORTS & SETUP ===
import sys

print("=" * 60)
print("IMPORTS & SETUP")
print("=" * 60)

# Clear any cached utils modules to fix circular import
for module in list(sys.modules.keys()):
    if module.startswith('utils'):
        del sys.modules[module]

# Ensure sys.path includes PROJECT_ROOT
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

# Import utilities
from utils.helpers import load_csv, set_seed
from utils.config import load_config
print("✓ Utils imported")

# Standard data science imports
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
print("✓ Libraries imported")

# Load config and set random seed
cfg = load_config()
set_seed(cfg['model']['random_state'])
print(f"✓ Seed set to {cfg['model']['random_state']}")

print("=" * 60)
```

```python
# === ANALYSIS: Load & Process Data ===
print("=" * 60)
print("DATA ANALYSIS")
print("=" * 60)

# Load data
df = load_csv('other/data.csv')
print(f"✓ Data loaded: {df.shape}")

# Filter data - High popularity items only
filtered_df = df[df['Popularity'] == 'High']
print(f"✓ Filtered data: {filtered_df.shape}")

# Prepare features and target
X = filtered_df[['Advertising Expenditure', 'Campaign Engagement Score', 'Product Price']]
y = filtered_df['Sales']

# Train linear regression
model = LinearRegression()
model.fit(X, y)

# Results
print(f"\n✓ Model trained successfully")
print(f"  Coefficients: {np.round(model.coef_, 4)}")
print(f"  Intercept: {model.intercept_:.2f}")
print(f"  R² Score: {model.score(X, y):.4f}")

print("=" * 60)
```
