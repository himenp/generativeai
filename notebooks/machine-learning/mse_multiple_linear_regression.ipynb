```python
#Step 1: import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import sys

# Set data path based on environment
if 'google.colab' in sys.modules:
    from google.colab import drive
    drive.mount('/content/drive')
    data_path = '/content/drive/My Drive/Projects/GenerativeAI/data/'
else:
    data_path = '../../data/'

#Step 2: read data
df = pd.read_csv(data_path + 'Sales.csv')

#Step 3: prepare data
X = df[['Product Price', 'Advertising Expenditure']]
y = df['Sales']

#Step 4: splitting the data in 70:30 ratio for train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X,    # specifying the independent variables
    y,    # specifying the dependent variable
    test_size=0.30,    # specifying the size of the test set as a fraction of the whole data
    random_state=1    # specifying a seed value to enable reproducible results
)

#Step 5: linear regression
model = LinearRegression()
lr = model.fit(X_train, y_train)

#Step 6: Predictions
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
```
